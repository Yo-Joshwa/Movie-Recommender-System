```python
# genres
# id
# keywords
# title
# overview
# cast
# crew 
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast' , 'crew']]
```

```python
movies.dropna()
```

```python
import ast

def convert(obj):
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    return [i for i in obj] 
```

```python
movies.loc[:, 'genres'] = movies['genres'].apply(convert)
movies
```

```python
movies.loc[:, 'keywords'] = movies['keywords'].apply(convert)
movies
```

```python
import ast

def convert_cast(text):
    L = []
    counter = 0

    if isinstance(text, str):
        try:
            data = ast.literal_eval(text)
        except Exception as e:
            print(f"Failed to evaluate: {text}, error: {e}")
            return []
    else:
        data = text

    for i in data:
        if isinstance(i, dict) and counter < 3:
            name = i.get('name', None)
            if name:
                L.append(name)
                counter += 1
        elif isinstance(i, str) and counter < 3:
            L.append(i)
            counter += 1

    return L


```

```python
movies = movies.dropna(subset=['cast']) 
movies = movies[movies['cast'].apply(lambda x: isinstance(x, (str, list)))]  
movies['cast'] = movies['cast'].apply(convert_cast)
movies
```

```python
import ast

def fetch_director(text):
    L = []
    
    if isinstance(text, list):
        data = text
    elif isinstance(text, str):  
        try:
            data = ast.literal_eval(text)  
        except (ValueError, SyntaxError):
            return [] 
    else:
        return []  

    for i in data:
        if isinstance(i, dict) and i.get('job') == 'Director':
            L.append(i.get('name'))
    
    return L
```

```python
movies.loc[:, 'crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
def collapse(L):
    L1 = []
    for i in L:
        if isinstance(i, dict): 
            value = i.get('name', '')  
        else:
            value = str(i)  
        
        L1.append(value.replace(" ", ""))  
    
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies.loc[:, 'genres'] = movies['genres'].apply(collapse)
movies.loc[:, 'keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: str(x).split() if isinstance(x, str) else [])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    movie = movie.lower().strip()
    
    titles = new['title'].str.lower().str.strip()
    
    if movie not in titles.values:
        print(f" Movie '{movie}' not found in dataset.")
        return
    
    index = titles[titles == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    print(f"✅ Recommendations for '{movie.title()}':")
    for i in distances[1:6]:
        print(new.iloc[i[0]]['title'])

        
```
